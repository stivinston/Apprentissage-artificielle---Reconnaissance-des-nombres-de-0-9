```python
import librosa
import librosa.display 
import IPython.display as ipd
import numpy as np
import glob
import matplotlib.pyplot as plt 
```

```python
file_path='DATA/1 champlain 3.wav'
```

```python
def feature_extraction(file_path):
    x, sample_rate=librosa.load(file_path)
    plt.plot(x, lw=1)
    mfcc = librosa.feature.mfcc(y=x, sr=sample_rate, n_mfcc=50)
    return mfcc
```

```python
mfcc1=feature_extraction(file_path)
```

```python
file_path='0 Josiane 1.m4a'
mfcc2=feature_extraction(file_path)
```

```python
mfcc1.shape, mfcc2.shape
```

```python
librosa.display.specshow(mfcc1)
```

```python
file_path='DATA/1 Champlain 1.wav'
y, sample_rate=librosa.load(file_path)
y_trim,_=librosa.effects.trim(y, top_db=20)
mfcc = librosa.feature.mfcc(y=y_trim, sr=sample_rate, n_mfcc=50)
plt.figure(figsize=[10,15])
plt.subplot(3,1,1)
plt.plot(y, lw=1)
plt.subplot(3,1,2)
plt.plot(y_trim, lw=1)
plt.subplot(3,1,3)
librosa.display.specshow(mfcc)
```
